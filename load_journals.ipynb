# Czasopisma

```python
d = pd.read_excel('data/wykaz_czasopism_2019.xlsx', 0, header=2)
d.columns = list(d.iloc[0,:5].values) + ['issn 2', 'e-issn 2', 'Punkty'] + list(d.columns[8:])
d = d.iloc[1:,1:]
```

```python
# filter categories
l = ['informatyka', 'matematyka', 'informatyka techniczna i telekomunikacja']
d = d[list(d.columns[:7]) + l]
d = d[d[l].notna().any(axis=1)]
d = d.reset_index(drop=True)
```

```python
# removing 2nd issn/title/e-issn if same
d.loc[(d['issn'] == d['issn 2']), 'issn 2'] = np.nan
d.loc[(d['Tytuł 1'] == d['Tytuł 2']), 'Tytuł 2'] = np.nan
d.loc[(d['e-issn'] == d['e-issn 2']), 'e-issn 2'] = np.nan

# fill 1st title/issn/e-issn with 2nd if 1st not present
d.loc[d['Tytuł 1'].isna(), 'Tytuł 1'] = d[d['Tytuł 1'].isna()]['Tytuł 2']
d.loc[d['issn'].isna(), 'issn'] = d[d['issn'].isna()]['issn 2']
d.loc[d['e-issn'].isna(), 'e-issn'] = d[d['e-issn'].isna()]['e-issn 2']
```

# Konferencje

```python
k1 = pd.read_excel('data/wykaz_czasopism_2019.xlsx', 1, header=2)
k1 = k1.iloc[:,1:3]
k1.columns = ['title', 'points_2019']
k1['title'] = k1['title'].str.extract('((^.*[^\)\s]\s*$)|(.*(?=\([^\)]*\)\s?$)))') # remove additional data in titles
k1['title'] = k1['title'].str.extract('(.*[^\s](?=\s*$))') # remove trailing spaces
```

```python
k1
```

```python
k2 = pd.read_excel('data/wykaz_czasopism_2021.xlsx', 1, header=0) # header was 2 for 2019 data
k2 = k2.iloc[:,1:3]
k2.columns = ['title', 'points_2021']
k2['title'] = k2['title'].str.extract('((^[^\[]*$)|(.*(?=\s\[)))') # remove more additional data in titles (present in 2021 data)
k2['title'] = k2['title'].str.extract('((^.*[^\)\s]\s*$)|(.*(?=\([^\)]*\)\s?$)))') # remove additional data in titles
k2['title'] = k2['title'].str.extract('(.*[^\s](?=\s*$))') # remove trailing spaces
```

```python
k2
```

```python
k = k2.set_index('title').join(k1.set_index('title'), how='outer')
k
```
